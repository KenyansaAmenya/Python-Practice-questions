18. Write a program that takes a list of numbers from the user (comma separated), converts them into a list of integers, and prints only the even numbers.

```python
# Get input from user
user_input = input("Enter numbers separated by commas: ")

# convert input string to list of integers
numbers = [int(x.strip()) for x in user_input.split(",")]

# Find even numbers
even_numbers = [num for num in numbers if num % 2 == 0]
print("Even numbers:", even_numbers)
```

another method

```python
# Get input from user
user_input = input("Enter numbers separated by commas: ")

numbers = [int(x.strip()) for x in user_input.split(",")]

print("Even numbers:")
for n in numbers:
    if n % 2 == 0:
        print(n)    
```

28. Write a program that counts how many times each word appears in a list of words: ["apple", "banana", "apple", "cherry", "banana", "banana"].

```python
words = ["apple", "banana", "apple", "cherry", "banana", "banana"]
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1  

print(word_count) 
```

35. Ask user to input three numbers and store them in a list.

 Then, convert the list to a set and check if any duplicates were removed; tell the user if there were duplicates.

```python

num1 = input("Enter first number: ")
num2 = input("Enter second number: ")
num3 = input("Enter third number: ")

numbers = [num1, num2, num3]

# Convert list to a set
unique_numbers = set(numbers)

# Check if duplicates were deleted
if len(unique_numbers) < len(numbers):
    print("There were duplicates.")
else:
    print("No duplicates.")
```
